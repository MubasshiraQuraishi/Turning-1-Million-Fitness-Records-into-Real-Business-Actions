```python
df1 = df[['age', 'gender', 'height_cm', 'weight_kg',
       'activity_type', 'duration_minutes', 'intensity', 'calories_burned',
       'avg_heart_rate', 'hours_sleep', 'stress_level', 'daily_steps',
       'hydration_level', 'bmi', 'resting_heart_rate',
       'blood_pressure_systolic', 'blood_pressure_diastolic',
       'health_condition', 'smoking_status', 'fitness_level']]
```

```python
num_cols = df1.select_dtypes(exclude='O').columns 
cat_cols = df1.select_dtypes(include='O').columns 

numeric_summary = df.groupby('participant_id')[num_cols].mean().reset_index()

def safe_mode(series):
    modes = series.mode()
    if not modes.empty:
        return modes.iloc[0]
    else:
        return None 

categorical_summary = df.groupby('participant_id')[cat_cols].agg(safe_mode).reset_index()


df_summarized = pd.merge(numeric_summary, categorical_summary, on='participant_id', how='left')
```

```python
df_summarized
```

```python
age_group= []
for i in df_summarized['age']:
    if i < 18:
        age_group.append("<18")
    if i <= 25:
        age_group.append('18-25')
    elif i <= 30:
        age_group.append('26-30')
    elif i <= 35:
        age_group.append('31-35')
    elif i <= 40:
        age_group.append('36-40')
    elif i <= 45:
        age_group.append('41-45')
    elif i <= 50:
        age_group.append('46-50')
    elif i <= 55:
        age_group.append('51-55')
    elif i <= 60:
        age_group.append('56-60')
    elif i <= 65:
        age_group.append('61-65')
    elif i > 65:
        age_group.append('65+')
        
df_summarized['age_group'] = age_group
```

```python
plt.figure(figsize=(15,5))
most_walks = df_summarized.groupby('age_group')['daily_steps'].sum()
sns.barplot(x = most_walks.index, y = most_walks.values)
plt.title("Walks by age-group")
plt.show()
```

```python
from scipy.stats import ttest_ind
male_stress = df_summarized[df_summarized['gender'] == "M"]['stress_level']
female_stress = df_summarized[df_summarized['gender'] == "F"]['stress_level']

tstat, pval = ttest_ind(male_stress, female_stress)

print(f"tstat : {tstat}")
print(f"P value : {pval}")

if pval < 0.05:
    print("Male and Female participants endures same amount of stress")
else:
    print("Male and Female participants endures different amount of stress")
    
print(f"Male stress level : {np.mean(male_stress)}") 
print(f"Female stress level : {np.mean(female_stress)}") 
```
